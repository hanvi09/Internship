# Splitting the data for training and testing:


```python
from sklearn.model_selection import train_test_split
```

```python
scale1=PowerTransformer(method='yeo-johnson')
x_scaled=pd.DataFrame(scale1.fit_transform(x),columns=x.columns)
```

```python
x_scaled
```

```python
x_train,x_test,y1_train,y1_test=train_test_split(x_scaled,y1,train_size=0.75, random_state=42, shuffle=True)
```

# Finding the best random state


```python
from sklearn.linear_model import  LinearRegression

LR=LinearRegression()

for i in range(0,100):
    x_train,x_test,y1_train,y1_test=train_test_split(x,y1,test_size=0.22,random_state=i)
    LR.fit(x_train,y1_train)
    LR_predict_train=LR.predict(x_train)
    LR_predict_test=LR.predict(x_test)
    print(f'At random state{i}, The training accuracy is :-{r2_score(y1_train,LR_predict_train)}')
    print(f'At random state{i}, The test accuracy is :-{r2_score(y1_test,LR_predict_test)}')
    print('\n')
```

# MODEL BUILDING

```python
from sklearn.linear_model import  LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

# MODEL SELECTION

```python
linear_model=LinearRegression()
rfr_model=RandomForestRegressor(max_depth=2,max_features="sqrt")
dtr_model=DecisionTreeRegressor(criterion='poisson',random_state=111)
```

# TRAINING THE MODEL

```python
linear_model.fit(x_train,y1_train)
rfr_model.fit(x_train,y1_train)
dtr_model.fit(x_train,y1_train)
```

# PREDICTION

```python
pred_linear=linear_model.predict(x_test)
pred_rfr=rfr_model.predict(x_test)
pred_dtr=dtr_model.predict(x_test)
```

# Evaluation Metrics


```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RandomizedSearchCV
from math import sqrt
```

```python
se_linaer=mean_squared_error(y1_test,pred_linear)
se_rfr=mean_squared_error(y1_test,pred_rfr)
se_dtr=mean_squared_error(y1_test,pred_dtr)
```

```python
rmse_linear=sqrt(se_linaer)
rmse_rfr=sqrt(se_rfr)
rmse_dtr=sqrt(se_dtr)
```

```python
print(f"RMSE of Linear Regression is: ", rmse_linear)
print(f"RMSE of Random Forest Regressor is: ", rmse_rfr)
print(f"RMSE of Decision Tree Regressor is: ", rmse_dtr)
```

```python
print(f"R2 Score for Linear Regression is:",r2_score(y1_test,pred_linear,multioutput='variance_weighted')*100)
print(f"R2 Score for Random Forest Regressor  is:",r2_score(y1_test,pred_rfr,multioutput='variance_weighted')*100)
print(f"R2 Score for Decision Tree Regressor  is:",r2_score(y1_test,pred_dtr,multioutput='variance_weighted')*100)
```
