```python


from bs4 import BeautifulSoup
import requests
import pandas as pd
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup = BeautifulSoup(page.content)
soup
#for i in soup.find_all(['h1','h2','h3','h4','h5','h6'])
head=[]
for i in soup.find_all(['h1','h2','h3','h4','h5','h6']):
    
    head.append([i.text,i.name])
head 

df = pd.DataFrame(head, columns=["Heading","Tagname"])
print(df)
```

```python
## 2) Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice)
# from https://presidentofindia.nic.in/former-presidents.htm and make data frame

from bs4 import BeautifulSoup
import requests
import pandas as pd
page = requests.get('https://presidentofindia.nic.in/former-presidents')
soup = BeautifulSoup(page.content)
soup

name=[]
for i in soup.find_all(['h3']):
    
    name.append(i.text.replace('\n',''))
 

term=[]
for j in soup.find_all(['h5']):
    
    term.append(j.text)


df = pd.DataFrame({"Former President":name,"Term of Office":term})
print(df)
```

# b)	Top 10 ODI Batsmen along with the records of their team andrating.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page = requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
soup = BeautifulSoup(page.text,"html.parser")
table = soup.find("table",class_="table")


b_rating = table.find("div",class_="rankings-block__banner--rating").text
batsman = table.find_all("td",class_="table-body__cell rankings-table__name name")[1:10]

bat = table.find("div",class_="rankings-block__banner--name-large").text
b = [bat]
for i in batsman:
    b.append(i.text.replace('\n',''))
teams = table.find_all("span",class_="table-body__logo-text")[1:10]

b_team= table.find("div",class_="rankings-block__banner--nationality").text.replace('\n','')
team=[b_team]
for i in teams:
    team.append(i.text)

ratings = table.find_all("td",class_="table-body__cell rating")[1:10]
rating = [b_rating]
for i in ratings:
    rating.append(i.text)
data = list(zip(b,team,rating))

df = pd.DataFrame(data, columns = ["Batsman Name","Team","Ratings"])
df
```

# c)	Top 10 ODI bowlers along with the records of their team andrating.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page = requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling")
soup = BeautifulSoup(page.text)
table = soup.find("table",class_="table rankings-table")

f_bowler = table.find("div",class_="rankings-block__banner--name-large").text
b_team= table.find("div",class_="rankings-block__banner--nationality").text.replace('\n','')
b_rating = table.find("div",class_="rankings-block__banner--rating").text
bowlers = table.find_all("td",class_="table-body__cell rankings-table__name name")[1:10]

bowler = [f_bowler]
for i in bowlers:
    bowler.append(i.text.replace('\n',''))

teams = table.find_all("span",class_="table-body__logo-text")[1:10]
team=[b_team]
for i in teams:
    team.append(i.text)

ratings = table.find_all("td",class_="table-body__cell rating")[1:10]
rating = [b_rating]
for i in ratings:
    rating.append(i.text)
data = list(zip(bowler,team,rating))

df = pd.DataFrame(data, columns = ["Bowler Name","Team","Ratings"])
df
```
